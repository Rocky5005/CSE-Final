```python
# import statements
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from main import file_locate
```

```python
# sets directory to access csv file
file_locate('data-analysis.ipynb')
```

```python
df = pd.read_csv('cleaned-framingham.csv')
df.info()
```

```python
female_count = df['male'].value_counts()[0]
male_count = df['male'].value_counts()[1]
values = [female_count, male_count]
labels = ['female', 'male']
sns.barplot(x=labels, y=values)
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.title('Distribution of Sex')
plt.show()

sns.histplot(df['age'])
plt.xlabel('Age')
plt.title('Distribution of Age')
plt.show()

sns.histplot(df['cigsPerDay'])
plt.xlabel('Cigarettes per Day')
plt.title('Distribution of Cigarettes')
plt.show()

sns.histplot(df['BPMeds'])
plt.xlabel('BPMeds')
plt.title('Prescence of BPMeds')
plt.show()

sns.histplot(df['prevalentStroke'])
plt.xlabel('Stroke Prevalence')
plt.title('Prevalence of Stroke in Family History')
plt.show()

sns.histplot(df['prevalentHyp'])
plt.xlabel('Hypertension Prevalence')
plt.title('Prevalence of Hypertension in Family History')
plt.show()

sns.histplot(df['totChol'])
plt.xlabel('Total Cholesterol(mhg/dL)')
plt.title('Distribution of Cholesterol')
plt.show()

sns.histplot(df['sysBP'])
plt.xlabel('Systolic Blood Pressure(mmHg)')
plt.title('Distribution of Systolic Blood Pressure')
plt.show()

sns.histplot(df['diaBP'])
plt.xlabel('Diastolic Blood Pressure(mmHg)')
plt.title('Distribution of Diastolic Blood Pressure')
plt.show()

sns.histplot(df['BMI'])
plt.xlabel('BMI')
plt.title('Distribution of BMI')
plt.show()

sns.histplot(df['heartRate'])
plt.xlabel('Heart Rate(Beats/Min)')
plt.title('Distribution of Heart Rate')
plt.show()

sns.histplot(df['glucose'])
plt.xlabel('Glucose Level(mg/dl)')
plt.title('Distribution of Glucose Levels')
plt.show()

sns.histplot(df['TenYearCHD'])
plt.xlabel('Prescence of Risk')
plt.title('Risk of CHD')
plt.show()


```

```python
corr_matrix = df.corr()
fig = plt.figure(figsize = (15, 15))
sns.heatmap(corr_matrix, annot = True)
```
